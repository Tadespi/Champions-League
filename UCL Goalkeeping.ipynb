```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import CSV
goal_keeper = pd.read_csv('../Champions_League_2022/Resources/goalkeeping.csv')
goal_keeper.head()
```

```python
# Total number of saves
total_saves = goal_keeper['saved'].sum()
print(f"Total number of saves: {total_saves}")
```

```python
# Top 5 players with the most saves
top_savers = goal_keeper[['player_name', 'saved']].sort_values(by='saved', ascending=False).head()
print("Top 5 players with the most saves:")
print(top_savers)
```

```python
# Bar chart of the top 5 players with the most saves
top_savers.plot(kind='bar', x='player_name', y='saved', legend=False)
plt.title('Top 5 Players with Most Saves')
plt.xlabel('Player Name')
plt.ylabel('Saves')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate average saves per match
goal_keeper['saves_per_match'] = goal_keeper['saved'] / goal_keeper['match_played']
# Top 5 players with the highest average saves per match
top_saves_per_match = goal_keeper[['player_name', 'saves_per_match']].sort_values(by='saves_per_match', ascending=False).head()
print("Top 5 players with the highest average saves per match:")
print(top_saves_per_match)
```

```python
# Bar chart for top 5 players with the highest average saves per match
top_saves_per_match.plot(kind='bar', x='player_name', y='saves_per_match', legend=False)
plt.title('Top 5 Players with Highest Average Saves per Match')
plt.xlabel('Player Name')
plt.ylabel('Average Saves per Match')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 5 players with the most clean sheets
top_clean_sheets = goal_keeper[['player_name', 'cleansheets']].sort_values(by='cleansheets', ascending=False).head()
print("Top 5 players with the most clean sheets:")
print(top_clean_sheets)
```

```python
# Bar chart for top 5 players with the most clean sheets
top_clean_sheets.plot(kind='bar', x='player_name', y='cleansheets', legend=False)
plt.title('Top 5 Players with Most Clean Sheets')
plt.xlabel('Player Name')
plt.ylabel('Clean Sheets')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 5 players with the most penalty saves
top_penalty_saves = goal_keeper[['player_name', 'saved_penalties']].sort_values(by='saved_penalties', ascending=False).head()
print("Top 5 players with the most penalty saves:")
print(top_penalty_saves)
```

```python
# Bar chart for top 5 players with the most penalty saves
top_penalty_saves.plot(kind='bar', x='player_name', y='saved_penalties', legend=False)
plt.title('Top 5 Players with Most Penalty Saves')
plt.xlabel('Player Name')
plt.ylabel('Penalty Saves')
plt.xticks(rotation=45)
plt.show()
```

```python
# Increase the size of the figure
plt.figure(figsize=(20, 7))

# Average saves per match by club
avg_saves_by_club = goal_keeper.groupby('club')['saved'].sum() / goal_keeper.groupby('club')['match_played'].sum()

# Bar chart for average saves per match by club
avg_saves_by_club.plot(kind='bar', color='salmon')
plt.title('Average Saves per Match by Club')
plt.xlabel('Club')
plt.ylabel('Average Saves per Match')
plt.xticks(rotation=45)
plt.show()
```
